```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import re
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("C:\\Users\\sumanyu singh\\Downloads\\nlp_int_project\\Restaurant_Reviews 1.tsv",delimiter='\t',quoting=3)
```

```python
docs=df.iloc[:,-2]
#initalize lemmatizer and stop words
lemmatizer=WordNetLemmatizer()
#Prepocessing function
def preprocess_text(text):
    #convert text to lowercase
    text=text.lower()
    #remove punctuation and special character
    text=re.sub(rf"[{string.punctuation}]"," ",text)
    #remove numbers
    text=re.sub(r'\d+'," ",text)
    #tokenize text 
    tokens=word_tokenize(text)
    #Lemmatize tokens
    tokens=[lemmatizer.lemmatize(word) for word in tokens]
    #rejoin tokens into a string string 
    return " ".join(tokens)

#Apply preprocessing to each documnent
corpus=[preprocess_text(doc) for doc in docs]
```

```python
#Intialize TfidfVectorizer
vectorizer=TfidfVectorizer()
print(corpus)
x=vectorizer.fit_transform(corpus).toarray()
print(x)
y=df.iloc[:,-1].values
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1500,test_size=0.2)
```

```python
from sklearn.svm import SVC
rf=SVC()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
print("accuracy",accuracy_score(y_test,y_pred)*100)
print(classification_report(y_test,y_pred))
```
